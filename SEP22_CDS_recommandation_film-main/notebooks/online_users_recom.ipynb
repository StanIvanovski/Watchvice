```python
#Librairies
import pandas as pd
import numpy as np
import sklearn.cluster
```

Data Loading from yhe results of the clustering

```python
mega_cluster_number=10
```

```python
number_cluster=100
```

```python
#dataframe with the rating and the cluster for each movies
dfratings=pd.read_csv('../data/dfratings_cluster.csv',index_col=False)
```

```python
#dataframe with the genome and the cluster for each movies

genome_top=pd.read_csv('../data/genome_cluster.csv',header=1)
```

```python
#numpy array with the centrer of the cluster coordinates and the cluster label
cluster_centroid=np.genfromtxt('../data/cluster_centroid.csv',delimiter=";")
```

```python
cluster_centroid.shape
```

```python
#movies  qui donne la correspondance des titre des films, leur genre avec leur movieId, et un coefficient d'age
dfmovies=pd.read_csv("../data/dfmovies_coef_age.csv", sep=',', index_col="movieId")
display(dfmovies.info())
display(dfmovies.head(5))
```

Check that the data loading is correct

```python
dfratings.head()
```

```python
dfratings=dfratings.dropna()
```

```python
#elimation de la ligne de NaN lié à la conversion en csv d'une double colonne
genome_top=genome_top.dropna()
```

genome_top.head()

```python
genome_top=genome_top.rename(columns={"Unnamed: 51":"label"})
```

```python
genome_top["tagId"]=genome_top["tagId"].astype(int)
```

```python
genome_top["label"]=genome_top["label"].astype(int)
```

```python
dfratings["cluster"]=dfratings["cluster"].astype(int)
```

```python
genome_top["mega_label"]=genome_top["label"]
```

```python
dfratings["mega_label"]=dfratings["cluster"]
```

```python
print(cluster_centroid[:])
```

```python
#creation of model clf ky Agglometrive Ward method with 10 clusters in order to obtain 10 mega cluster
clf = sklearn.cluster.AgglomerativeClustering(n_clusters = mega_cluster_number, affinity = 'euclidean', linkage = 'ward')
```

```python
clf.fit(cluster_centroid)
```

```python
print(clf.labels_)
```

```python
print(len(clf.labels_))
```

```python
print(clf.labels_[27])
```

```python
dfratings["centroid"]=dfratings["cluster"]

for i in range(len(clf.labels_)) :
    genome_top.mega_label[genome_top["label"]==i]=clf.labels_[i]
    dfratings.mega_label[dfratings["cluster"]==i]=clf.labels_[i]
    #dfratings.centroid[dfratings["cluster"]==i]=cluster_centroid[i]
```

```python
dfratings_by_movie=pd.DataFrame(dfratings.groupby("movieId").mean())

```

```python
dfratings_by_movie.head()
```

```python
popularity=dfratings.groupby("movieId").count()
popularity.head()
```

```python
dfratings_by_movie["popularity"]=popularity["userId"]
dfratings_by_movie=dfratings_by_movie.drop(["Unnamed: 0", "userId", "timestamp"], axis=1)
```

```python
dfratings_by_movie.head()
```

```python
dfratings_by_movie=dfratings_by_movie.merge(right=dfmovies,on='movieId',how='left')
```

```python
dfratings_by_movie=dfratings_by_movie.drop(["genres", "year"], axis=1)
dfratings_by_movie.head()
```

```python
dfratings_by_movie["coef_age"]=dfratings_by_movie["rating"]+20*dfratings_by_movie["coef_age"]
```

```python
dfratings_by_movie.to_csv("../data/mega_cluster_by_movie.csv")
```
